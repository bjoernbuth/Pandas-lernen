```python
## Dataframes nebeneinander anzeigen (Versuch)

# from IPython.display import display, HTML

# CSS = """
# .output {
#     flex-direction: row;
# }
# """

# HTML('<style>{}</style>'.format(CSS))

import numpy as np
import pandas as pd

class my_display(object):
    """Display HTML representation of multiple objects"""
    template = """<div style="float: left; padding: 10px;">
    <p style='font-family:"Courier New", Courier, monospace'>{0}</p>{1}
    </div>"""

    def __init__(self, *args):
        self.args = args

    def _repr_html_(self):
        return '\n'.join(self.template.format(a, eval(a)._repr_html_())
                     for a in self.args)

    def __repr__(self):
       return '\n\n'.join(a + '\n' + repr(eval(a))
                       for a in self.args)
```

## Erstes Beispiel fuer semi-joins

```python
df5 = pd.DataFrame(np.random.randint(0,4,size=(6, 2)), columns=list('AB'))
df6 = pd.DataFrame(np.random.randint(0,4,size=(6, 2)), columns=list('BC'))
my_display('df5','df6')
```

```python
df5['B'].isin(df6['B'])
```

```python
## Alte Spalten von links bei denen 'B' auch rechts vorkommen
df_temp = df5[df5['B'].isin(df6['B'])]
my_display('df5','df6', 'df_temp')
```

```python
## Alte Spalten von links bei denen 'B' rechts nicht vorkommen
df_temp = df5[~ df5['B'].isin(df6['B'])]
my_display('df5','df6', 'df_temp')
```

```python
df5.merge(df6, how='left', indicator=True)
```

## From Stackoverflow

https://stackoverflow.com/questions/13269890/cartesian-product-in-pandas

```python
import numpy as np 
import pandas as pd 
import itertools

def cartesian(df1, df2):
    rows = itertools.product(df1.iterrows(), df2.iterrows())

    df = pd.DataFrame(left.append(right) for (_, left), (_, right) in rows)
    return df.reset_index(drop=True)
```
